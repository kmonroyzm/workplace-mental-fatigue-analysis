```python

from google.colab import drive
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
file_path = '/content/drive/MyDrive/python_cursos/employee_survey.csv'
df = pd.read_csv('/content/drive/MyDrive/python_cursos/employee_survey.csv', encoding='ISO-8859-1')
```

```python
df.columns = [
    "ID_del_empleado",
    "Fecha_de_incorporacion",
    "Genero",
    "Tipo_de_trabajo",
    "Configuracion_forma_de_trabajo",
    "Designacion",
    "Asignacion_de_horas",
    "Puntuacion_fatiga_mental",
    "Sueldo"
]
```

```python
nulos = df.isnull().sum()
print("Elementos nulos en cada columna:")
print(nulos)
```

```python
df.rename(columns={
    "ID_del_empleado": "Empleado_Clave",
    "Fecha_de_incorporacion": "Fecha_Contrato",
    "Genero": "Genero_Empleado",
    "Puntuacion_fatiga_mental": "Fatiga_mental"
}, inplace=True)

# Mostrar las primeras filas para verificar
print(df.head())
```

```python
# Eliminar la columna Sueldo
df.drop(columns=["Sueldo"], inplace=True)

# Mostrar las primeras filas para verificar
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Graficar gráfico de barras de género y sueldo
df['Genero_Empleado'] = df['Genero_Empleado'].replace({'Female': 'Mujer', 'Male': 'Hombre'})  # Traducir géneros si es necesario

df_barras = df.groupby('Genero_Empleado')['Sueldo'].mean()

df_barras.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Promedio de Sueldo por Género')
plt.xlabel('Género del Empleado')
plt.ylabel('Sueldo Promedio')
plt.show()
```

```python
# Graficar gráfico de dispersión de sueldo vs. fatiga mental
plt.scatter(df['Fatiga_mental'], df['Sueldo'], color='purple')
plt.title('Sueldo vs Puntuación de Fatiga Mental')
plt.xlabel('Puntuación de Fatiga Mental')
plt.ylabel('Sueldo')
plt.show()
```
